Split the data into Training data and Testing data

By splitting the data into training and testing sets, you can train a machine learning model on the training data and evaluate its performance on the testing data. This helps to prevent overfitting and provides a more accurate estimate of the model's performance on unseen data.

where, => 'train_test_split' is a function from Scikit-learn's model_selection module that splits arrays or matrices into random train and test subsets. x and y are the input arrays to be split. In this case, x contains the feature variables, and y contains the target variable.

=> 'test_size = 0.2' specifies that 20% of the data should be used for testing, and the remaining 80% should be used for training.

=> 'stratify = y' ensures that the split is stratified, meaning that the proportion of each class in the target variable y is preserved in both the training and testing sets. This is important when working with classification problems, where the class distribution is not uniform.

=>'random_state = 2' sets the seed for the random number generator, ensuring that the split is reproducible.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size =0.2, stratify = y, random_state= 2)
```

The KNeighborsClassifier is a simple yet powerful supervised learning algorithm used for classification problems. It's a type of instance-based learning, where the classifier makes predictions based on the similarity between the input data and the training data.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error = []
for i in range(1, 12):
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    error.append(np.mean(y_test != y_pred))
```

By calling plt.plot, you're creating a line plot with the specified marker and color. The resulting plot will show the error values at each point in the range from 1 to 12, with a yellow star marker at each data point.

This code will generate a plot with the x-axis labeled as "Value", the y-axis labeled as "Error", and a title of "Error Plot". The plt.show() function is used to display the plot.

```python
# plot

plt.xlabel('k value')
plt.ylabel('error')

plt.plot(range(1,12), error, marker = '*', mfc = 'yellow')
plt.grid();
```

In this case, the KNN model will find the 3 most similar neighbors for each sample in the test data and use their labels to make a prediction. The predicted labels are stored in y_pred, which can be used to evaluate the model's performance.

```python
knn = KNeighborsClassifier(n_neighbors= 3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```
